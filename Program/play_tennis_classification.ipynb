## Play Tennis Classification

### Import Dataset

```python
# Import dataset berbentuk csv
import pandas as pd
data = pd.read_csv("play_tennis.csv")
data
```

### Drop Kolom

```python
# Menghapus fitur 'day'
data.drop('day', axis=1, inplace=True)
data
```

### Label Encoding

```python
# Mengubah data kategorikal menjadi numerikal 
from sklearn.preprocessing import LabelEncoder
Le = LabelEncoder()
data['outlook'] = Le.fit_transform(data['outlook'])
data['temp'] = Le.fit_transform(data['temp'])
data['humidity'] = Le.fit_transform(data['humidity'])
data['wind'] = Le.fit_transform(data['wind'])
data['play'] = Le.fit_transform(data['play'])
data
```

### Feature Extraction

```python
# Mengambil 4 kolom awal untuk melatih model 
x = data.iloc[:,:4].values
x
```

### Label Extraction

```python
# Mengambil kolom terakhir untuk menjadi label prediksi 

y = data.iloc[:,-1].values
y
```

### Data Splitting

```python
# Memisahkan data menjadi data test dan train 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.7)
```

### Training & Evaluation

```python
from sklearn import naive_bayes
from sklearn import metrics

# Melatih model Naive Bayes (BernoulliNB) dengan data training
nb = naive_bayes.BernoulliNB()
nb.fit(x_train, y_train)

# Melakukan prediksi pada data uji
y_pred = nb.predict(x_test)

# Mengevaluasi performa model dengan menghitung akurasi
score = metrics.accuracy_score(y_test, y_pred)
print("Akurasi prediksi: ", score)

# Menampilkan classification report (Precision, Recall, F1-Score)
print("\nClassification Report:")
print(metrics.classification_report(y_test, y_pred))
```

### Save Model

```python
# Menyimpan model yang sudah dilatih
import pickle
import pandas as pd
pd.to_pickle(nb, "model_play_tennis.pkl")
```
